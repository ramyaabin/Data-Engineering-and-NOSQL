### SQL Coding Assignment on SQLite

```python
import sqlite3
import pandas as pd
```

```python
# connect to SQLiteDB & sample query
conn = sqlite3.connect("northwind.db")
pd.read_sql("SELECT * FROM Categories limit 5;", conn)
```

### 1 Write a query to get Product name and quantity/unit for products with more than 50 units in stock

```python
pd.read_sql("SELECT ProductName, UnitsInStock FROM products where UnitsInStock >50 order by UnitsInStock desc;", conn)
```

### 2 Write a query to get most expensive and least expensive Product list (name and unit price)

```python
pd.read_sql(("SELECT ProductName, UnitPrice FROM Products ORDER BY UnitPrice DESC LIMIT 1"),conn)
```

```python
pd.read_sql(("SELECT ProductName, UnitPrice FROM Products ORDER BY UnitPrice ASC LIMIT 1"),conn)
```

### 3 Write a query to get Product list (name, unit price) of ten most expensive products

```python
pd.read_sql(("SELECT ProductName, UnitPrice FROM Products ORDER BY UnitPrice DESC LIMIT 10"),conn)
```

### 4 Write a query to count current and discontinued products

```python
pd.read_sql(('''SELECT 
    SUM(CASE WHEN Discontinued = 0 THEN 1 ELSE 0 END) AS 'Current Products',
    SUM(CASE WHEN Discontinued = 1 THEN 1 ELSE 0 END) AS 'Discontinued Products'
FROM Products;'''),conn)
    
```

### 5 Write a query to get productname, unit price, quantity per unit, units in stock and category for all discontinued products

```python
pd.read_sql('''SELECT ProductName, UnitPrice, QuantityPerUnit, UnitsInStock,
CategoryName FROM products p left join Categories c
on p.CategoryID = c.CategoryID where discontinued = 1;''', conn)
```

### 2.Total number of products for each category

```python
pd.read_sql(('''SELECT 
    CategoryName, COUNT(ProductID) as 'Total Products'
FROM Products
JOIN Categories
ON Products.CategoryID = Categories.CategoryID
GROUP BY CategoryName'''),conn)
```

### 3. Top 5 categories with the highest average unit price

```python
pd.read_sql(('''SELECT 
    CategoryName, AVG(UnitPrice) as 'Avg Unit Price'
FROM Products
JOIN Categories
ON Products.CategoryID = Categories.CategoryID
GROUP BY CategoryName
ORDER BY AVG(UnitPrice) DESC
LIMIT 5'''),conn)
```
