```python
# connect to SQLiteDB & sample query
conn = sqlite3.connect("northwind.db")
pd.read_sql("SELECT * FROM Categories limit 5;", conn)
```

#### 1 Write a query to get Product name and quantity/unit for products with more than 50 units in stock

```python
pd.read_sql("select productname,quantityperunit from products where unitsinstock>50;",conn)
```

#### 2 Write a query to get most expensive and least expensive Product list (name and unit price)

```python
#most expensive
pd.read_sql("select productname,unitprice from products order by unitprice desc limit 1;",conn)
```

```python
#Least Expensive
pd.read_sql("select productname,unitprice from products order by unitprice asc limit 1;",conn)
```

#### 3 Write a query to get Product list (name, unit price) of ten most expensive products. 

```python
#10 most expensive products
pd.read_sql("select productname,unitprice from products order by unitprice desc limit 10;",conn)
```

#### 4 Write a query to count current and discontinued products

```python
pd.read_sql("select sum(case when discontinued =1 then 1 end) as Discontinued,sum(case when discontinued =0 then 1 end) as Current from  products;",conn)
```

#### 5 Write a query to get productname, unit price, quantity per unit, units in stock and category for all discontinued products

```python
pd.read_sql("select productname,unitprice,quantityperunit,unitsinstock from products where discontinued=1;",conn)
```
